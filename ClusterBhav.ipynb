```python
import pandas as pd
import numpy as np
#import yfinance as yf
from ta import add_all_ta_features
from ta.utils import dropna
from ta.volatility import BollingerBands
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv('/content/bhavcopy2022July.csv')
```

```python
data=data[data['SERIES']=='EQ'].reset_index(drop=True)
```

```python
start=data[data['TIMESTAMP']=='03-JAN-2022']
start
```

```python
data
```

```python
pd.DataFrame(data.groupby('SYMBOL')['OPEN'].mean()).reset_index()
```

```python
ahead=10
import warnings
warnings.filterwarnings("ignore")
cluster_rat=0.1
perfs=[]
for company in start.head(100)['SYMBOL'].unique():
  print(company)
  dat=data[data['SYMBOL']==company]
  dat.reset_index(inplace=True,drop=True)
  
  indicator_bb = BollingerBands(close=dat["CLOSE"], window=20, window_dev=2)

  # Add Bollinger Bands features
  dat['bb_bbm'] = indicator_bb.bollinger_mavg()
  dat['bb_bbh'] = indicator_bb.bollinger_hband()
  dat['bb_bbl'] = indicator_bb.bollinger_lband()

  # Add Bollinger Band high indicator
  dat['bb_bbhi'] = indicator_bb.bollinger_hband_indicator()

  # Add Bollinger Band low indicator
  dat['bb_bbli'] = indicator_bb.bollinger_lband_indicator()
  for i in ['bb_bbm','bb_bbh','bb_bbl']:
    dat[i]=dat[i]/dat['CLOSE']

  
  dat['gain']=1
  for i in range(len(dat)):
    try:
      dat.loc[i,'gain']=dat.loc[i+ahead,'CLOSE']/dat.loc[i,'CLOSE']
    except:
      pass

  dat=dat.dropna().reset_index(drop=True)

  df=dat.head(100)
  if len(dat)>100:
    test=df.tail(len(dat)-100)

    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(df[['bb_bbm','bb_bbh','bb_bbl','bb_bbhi','bb_bbli']])
    kmeans = KMeans(
            init="random",
              n_clusters=int(len(df)*cluster_rat),
              n_init=10,
            max_iter=300,
            
              )
    
    kmeans.fit(scaled_features)
    df['cluster']=kmeans.labels_
    x=pd.DataFrame(df.groupby('cluster')['gain'].mean()).reset_index().sort_values('gain',ascending=False).reset_index(drop=True)
    bestcluster=x.loc[0,'cluster']

    test_scaled_features = scaler.transform(test[['bb_bbm','bb_bbh','bb_bbl','bb_bbhi','bb_bbli']])
    test['cluster']=kmeans.predict(test_scaled_features)
    sub=test[test['cluster']==bestcluster]
    if len(sub)>0:
      #print(company)
      print(np.mean(sub['gain']))
      print(sub)


  
```
